```python
import sympy as sp
import numpy as np
x, xc, y, yc, a, b, c, r_obst = sp.symbols('x, xc, y, yc, a, b, c, r_obst')
r  = sp.sqrt((x-xc)**2+(y-yc)**2)  # used for short-hand when defining B

# a: function of obstacle radius
# b: argument of exponential (we chose 100)
# c: cost at obstacle diameter (we chose 2)
a = r_obst**8*(c+r_obst**4*sp.exp(-b*r_obst**2))
B = (a/r**8)-(r**4)*sp.exp(-b*(r)**2)
```

```python
B
```

```python
print(B)
```

```python
print(B.diff(x))
```

```python
print(B.diff(x).diff(x))
```

```python
print(B.diff(y))
```

```python
print(B.diff(y).diff(y))
```

```python
print(B.diff(x).diff(y))
```
